```python
import numpy as np
import matplotlib.pyplot as plt
import CoolProp.CoolProp as CP
import scipy.optimize as opt
%matplotlib inline
plt.style.use('ggplot')
```

```python
P = 100000

def sistema(v):
    T = v
    p_sat_1 = CP.PropsSI('P','T',T,'Q',0,'Hexane')
    p_sat_2 = CP.PropsSI('P','T',T,'Q',0,'Heptane')
    return(P - p_sat_2 - x_1 * (p_sat_1 - p_sat_2))

x = np.linspace(0,1,100)
temp, y = [], []
guess = 371
for x_1 in x:
    solv =  opt.fsolve(sistema, guess)
    temp.append(solv)
    guess = solv
    y.append((1 - CP.PropsSI('P','T',solv,'Q',0,'Heptane') / P) / \
             (1 - CP.PropsSI('P','T',solv,'Q',0,'Heptane')/CP.PropsSI('P','T',solv,'Q',0,'Hexane')) )
temp = np.array(temp) - 273.15
```

```python
fig = plt.figure(figsize=(7,8))
fig.text(0.42, 0.40, "L", fontsize=20, color="k")
fig.text(0.48, 0.50, "V + L", fontsize=20, color="k")
fig.text(0.62, 0.60, "V", fontsize=20, color="k")
ax = fig.add_subplot(1,1,1)
ax.plot(x,temp,label=r"Ponto de bolha")
ax.plot(y,temp,label=r"Ponto de orvalho")
ax.legend(loc=1)
ax.set_xlim(0,1)
ax.set_xlabel(r'$x_1,y_1$ (-)',fontsize=18)
ax.set_ylabel(r'$T$ (°C)',fontsize=18)
ax.set_title('Equílibrio Líquido-Vapor\nMistura n-hexano e n-pentano')
```
