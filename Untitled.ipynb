```python
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from matplotlib import pyplot as plt

fig = plt.figure()
ax = Axes3D(fig)

alfa = 5.
beta = 1.1

x = np.linspace(0,alfa, 10)
y = np.linspace(-beta, beta, 10)

def polio(xees):
    a=xees**2
    b=a*xees
    print(2*b - 3 *a+1)
    return 2*b - 3 *a+1

def halfcell(a, b,xx,yy):
    return polio(xx/a)*polio(yy/b)


X,Y = np.meshgrid(x,y)
Z = halfcell(alfa, beta, X,Y)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')
plt.show()
```
